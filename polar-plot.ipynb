```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

r = np.arange(0, 2, 0.01)

theta = np.linspace(0, 2 * np.pi, 1000)

w0 = 1
w1 = 0.5
s2 = 1

ws1 = [1, 0.5, 1]
ws2 = [1, 2, -1]


def G(theta, ws):
    return np.sum([w * np.exp(1j * i * np.pi * np.sin(theta)) for i, w in enumerate(ws)], axis=0)



print(type(G(np.pi, ws1)))



fig, ax = plt.subplots( subplot_kw={'projection': 'polar'}, figsize=(10, 10))
ax.plot(theta, np.abs(G(theta, ws1)), color='b')
ax.plot(theta, np.abs(G(theta, ws2)), color='r')

# ax.plot(theta, [ c.imag for c in G(theta, ws1)], color='r')
# ax.plot(theta, G(theta, ws2), color='g')
# ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')

# # ax[1].plot(theta, G(theta, ws2), color='g')
# # ax[1].set_rmax(2)
# ax[1].set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
# ax[1].set_rlabel_position(-22.5)  # Move radial labels away from plotted line

# ax[1].grid(True)

# ax[0].set_title("A line plot on a polar axis", va='bottom')
plt.show()
```
